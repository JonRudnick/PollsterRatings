```python
#author: Jonathan Rudnick
#Final Project
#12/16/15
import pandas
from sklearn import linear_model as lm
from sklearn import cross_validation as cv
from sklearn import metrics

poll_data = pandas.read_csv('PollData.csv')
#print (poll_data)

poll_data = poll_data.drop('Num', axis = 1)
poll_data = poll_data.drop('Pollster', axis = 1)
poll_data = poll_data.drop('Bias', axis = 1)
poll_data = poll_data.drop('Mean-Reverted Bias', axis = 1)
poll_data = poll_data.drop('House Effect', axis = 1)
poll_data['Database']=poll_data['Database'].replace('No',0)
poll_data['Database']=poll_data['Database'].replace('Yes',1)
poll_data['Calls Cellphones?']=poll_data['Calls Cellphones?'].replace('No',0)
poll_data['Calls Cellphones?']=poll_data['Calls Cellphones?'].replace('Sometimes',0.5)
poll_data['Calls Cellphones?']=poll_data['Calls Cellphones?'].replace('Yes',1)
poll_data['Banned by 538']=poll_data['Banned by 538'].replace('No',0)
poll_data['Banned by 538']=poll_data['Banned by 538'].replace('Yes',1)
poll_data = pandas.get_dummies(poll_data)
#print (poll_data)

train_array = []
test_array = []
epoch_array = []
(poll_train, poll_test) = cv.train_test_split(poll_data, test_size = 0.2)
#print (poll_data)

predictors = poll_data.columns[0:-1]
for i in range (50,100,5):
    perc_alg = lm.SGDClassifier(eta0 = 0.0000001, n_iter = i)
    perc_alg.fit(poll_train[predictors],poll_train["538 Grade_A"])
    poll_test_preds = perc_alg.predict(poll_test[predictors])
    poll_train_preds = perc_alg.predict(poll_train[predictors])
    train_array.append(metrics.accuracy_score(poll_train["538 Grade_A"], poll_train_preds))
    test_array.append(metrics.accuracy_score(poll_test["538 Grade_A"], poll_test_preds))
    epoch_array.append(i)
print (metrics.accuracy_score(poll_test["538 Grade_A"], poll_test_preds))

%matplotlib inline
import matplotlib.pyplot as plt
plt.plot(epoch_array,train_array, 'ro-', label = "Training Set")
plt.plot(epoch_array,test_array, 'bs-', label = "Test Set")
plt.legend(loc = 'lower right', shadow = True)
plt.show()
```
